## Mine

```python
def f1(x):
    return np.sqrt(x)
def f2(x):
    return x**3
def f3(x):
    return 2**x
def f4(x):
    return 1 / (x + 1)

#generate x-values
x_values = np.linspace(-10, 10, 100)

#generate y-values using the functions
y_values1 = f1(x_values)
y_values2 = f2(x_values)
y_values3 = f3(x_values)
y_values4 = f4(x_values)

#create and label 4 separate charts
plt.figure(figsize=(8, 6))

#chart 1
plt.plot(x_values, y_values1, label=r'$y = /squrt{x}$') #label here is used for the legend
plt.title('Square Root Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

#chart 2
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values2, label=r'$y = x^3$') #r' indicates a raw string for LaTeX commands
plt.title('Cubic Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

#chart 3
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values3, label=r'$y = 2^x$')
plt.title('Exponential Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

#chart 4
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values4, label=r'$y = \frac{1}{x+1}$')
plt.title('Reciprocal Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

## 3. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
# re-use functions above

#generate x-values
x_values = np.linspace(0, 10, 25)

#generate y-values using the functions
y_values1 = f1(x_values)
y_values2 = f2(x_values)
y_values3 = f3(x_values)
y_values4 = f4(x_values)

#create and label 4 separate charts
plt.figure(figsize=(12, 8))

#chart 1
plt.subplot(2,2,1)
plt.plot(x_values, y_values1, label=r'$y = /squrt{x}$', marker='o') #label here is used for the legend
plt.title('Square Root Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

#chart 2
plt.subplot(2,2,2)
plt.plot(x_values, y_values2, label=r'$y = x^3$', marker='^') #r' indicates a raw string for LaTeX commands
plt.title('Cubic Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

#chart 3
plt.subplot(2,2,3)
plt.plot(x_values, y_values3, label=r'$y = 2^x$', ls='--')
plt.title('Exponential Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

#chart 4
plt.subplot(2,2,4)
plt.plot(x_values, y_values4, label=r'$y = \frac{1}{x+1}$', marker='*', ls=':')
plt.title('Reciprocal Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

#adjust layout to prevent overlapping of labels
plt.tight_layout()

plt.show()
```
